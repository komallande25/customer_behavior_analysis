```python
import pandas as pd
df =pd.read_csv('/content/customer_shopping_behavior.csv')
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df = df.rename(columns={"purchase_amount_(usd)" : "purchase_amount"})
```

```python
df.info()
```

```python
#creating a column age_group
labels = ['Young' , 'Adult' , 'Middle_aged' , 'senior']
df['age_group'] = pd.qcut(df['age'], q= 4 ,labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
#create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly' : 14 ,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90 ,
    'Bi-Weekly' : 14 ,
     'Annually' : 365 ,
    'Every 3 Months' : 90
}

df['parchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['parchase_frequency_days' , 'frequency_of_purchases']].head()
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
df = df.drop('promo_code_used', axis =1)
```
